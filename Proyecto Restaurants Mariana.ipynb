```python
import pandas as pd
import numpy as np
```

```python
rest = pd.read_csv('TA_restaurants_curated.csv',encoding='utf8', index_col=0)
rest.info()
```

```python
print(len(rest))
rest.head()
```

```python
pd.options.display.float_format = '{:.2f}'.format
rest.describe()
```

```python
rest['Number of Reviews'] = rest['Number of Reviews'].fillna(0)
rest['Number of Reviews'] = rest['Number of Reviews'].round(0).astype('int')
rest = rest.drop_duplicates('ID_TA', keep='first')
rest.Rating.replace(-1, 0, inplace=True)
```

```python
rest.info()
rest.dtypes
```

```python
rest.rename(columns={'Name': 'name',
            'City': 'city',
            'Ranking': 'ranking',
            'Rating': 'rating',
            'Reviews': 'reviews',
            'Cuisine Style':'cuisine_style',
            'Price Range':'price_range',
            'Number of Reviews':'reviews_number'}, inplace=True)
```

```python
rest.head()
```

```python
print("Número de Ciudades en el df: ", rest.city.nunique())
print("Número de Restaurants en el df: ", rest.ID_TA.nunique())
```

```python
df = rest.copy()
```

### TOP 10 Ciudades con más Restaurants

```python
top_cities = (df.groupby('city')['name'].count().to_frame()).sort_values(by='name',ascending = False).head(n=10)
top_cities2 = list(top_cities.index)
top_cities2
```

```python
df_filtered = df[df['city'].isin(top_cities2)]
df_filtered
```

```python
# Función MAP para evaluar el rango de precios en categorías.

df_filtered['price_range'] = df_filtered['price_range'].fillna('NA')
price_ranges = {'$': 'Barato', '$$ - $$$': 'Medio', '$$$$': 'Costoso', 'NA': 'SinInformacion'}
df_filtered['price_range'] = df_filtered.price_range.map(price_ranges)
df_filtered
```

```python
rango_precio = df_filtered.groupby(['city','price_range'])[['name']].count()
rating_mean = df_filtered.groupby(['city','price_range'])[['rating']].mean()
```

```python
rango_rating = rango_precio.join(rating_mean, rsuffix = ('_precio'))
rango_rating.rename(columns = {'name' : 'num_rest', 'rating':'rating_promedio'}, inplace = True)
rango_rating.sort_index()
```

```python
cocinas = df_filtered.loc[:, ['city', 'name', 'cuisine_style', 'rating', 'reviews_number', 'reviews','price_range']]
#Elimino los corchetes
cocinas.cuisine_style = cocinas.cuisine_style.str.replace('[', '')
cocinas.cuisine_style = cocinas.cuisine_style.str.replace(']', '')
cocinas.columns = ['city', 'place', 'cuisine_style', 'rating', 'reviews_number', 'reviews','price_range']
#Filtro las cocinas por aquellas que tengan su rating > a 5 y numero de reviews superior a la media
cocinas = cocinas[(cocinas['reviews_number'] >= cocinas.reviews_number.mean())&(cocinas['rating'] >= 5)]
cocinas
```

```python
cocinas['city2'] = cocinas['city']
cocinas2 = cocinas.set_index('city')
cocinas2
```

```python
#Maximo numero de reviews en esta lista
reviews = cocinas.groupby(['city'])[['city2','price_range','reviews_number']].max()
reviews
```

```python
tmp2 = (pd.merge(reviews, cocinas2, on='reviews_number', how='left')).sort_values(by = 'reviews_number',ascending = False)
tmp2 = tmp2.set_index(np.arange(len(tmp2.index)))
tmp2

tmp2.rename(columns={'city2_x': 'Ciudad',
            'place': 'Restaurant Sugerido',
            'price_range_x': 'Rango de precio',
            'cuisine_style': 'Estilo de Cocina',
            'reviews_number': 'Numero de Reviews'}, inplace=True)
tmp2
```

### Dónde es seguro comer:
- Los restaurants a continuación, tienen un promedio de **5** en su rating, todos tienen más comentarios de la media de comentarios de estas ciudades, por lo que serán una **apuesta segura** a la hora de visitarles! 
- Se aprecia en su columna de de reviews un sentimiento positivo por parte de los comensales y todos estos rest se encuentran en un rango medio de precio.
